<font size="5">Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. 
(2.5 points)</font>

```python
import pandas as pd
#Reading CSV and skipping row 0, now the dataframe has correct column names
df = pd.read_csv("Laliga.csv", skiprows=[0])
#Replacing '-' with 0
df.replace('-', 0, inplace=True)
```

```python
df.head()
```

---------------------------------------------------------------------------------------------------------------------
<font size="5">Print all the teams which have started playing between 1930-1980 (5 points)</font>

```python
#Debut column contains some Year in the format 2019 and some other in format 2019-20.
#Using split function to split at '-' and then keeping only the first part of the split
df['Debut'] = df['Debut'].str.split('-').str[0]
#Convert Debut columns from 'object' datatype to 'int64' for arithematic operation
df.Debut = df.Debut.astype('int64')
```

```python
#Conditional selecting of rows where Debut is from 1930 to 1980 inclusive. Assign a new dataframe
df_strtplay = df[(df['Debut'] >= 1930) & (df['Debut'] <=1980)]
#Print the team values
print(df_strtplay['Team'])
```

---------------------------------------------------------------------------------------------------------------------
<font size="5">Print the list of teams which came Top 5 in terms of points (2.5 points)</font>

```python
#Convert Points columns from 'object' datatype to 'int64' for arithematic operation
df.Points = df.Points.astype('int64')
```

```python
# Using nlargest function available in pandas 0.17.0b version. Assign it to new dataframe
df_toppoints = df.nlargest(5, 'Points')
# Print the team names and points of dataframe
print(df_toppoints[['Team','Points']])
```

```python
# Alternate way to do this
print(df.sort_values('Points', ascending=False).head()[['Team','Points']])
```

<font size="5">Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst</font>

```python
df.head(2)
#Converting GoalsFor column to int64 datatype
df.GoalsFor = df.GoalsFor.astype('int64')

#Converting GoalsAgainst column to int64 datatype
df.GoalsAgainst = df.GoalsAgainst.astype('int64')
```

```python
#Function returning a pandas series with Team names and their Goal Differences
Goal_diff_count = lambda x:pd.Series([x.Team, x.GoalsFor - x.GoalsAgainst], index=['Team', 'GoalDiff'])

#Applying the function to the data frame and printing it
print(df.apply(Goal_diff_count, axis=1, result_type='expand'))

#Using the same function to and finding team with Maximum Goal Difference
print('Team with maximum goal difference\n',df.apply(Goal_diff_count, axis=1, result_type='expand').max())

#Using the same function to and finding team with Minimum Goal Difference
print('Team with minimum goal difference\n',df.apply(Goal_diff_count, axis=1, result_type='expand').min())
```

<font size="5">Create a new column with name “Winning Percent” and append it to the data set (5 points) Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage</font>

```python
#Converting GoalsAgainst column to int64 datatype
df.GamesWon = df.GamesWon.astype('int64')

#Converting GoalsAgainst column to int64 datatype
df.GamesPlayed = df.GamesPlayed.astype('int64')

df.head(2)
```

```python
#Creating new column with winning percentage and adding it to dataframe.
#No numerical error as both Games won and Games played were converted to int64
df['Winning Percent'] = (df['GamesWon'] / df['GamesPlayed'])*100

#Printing top 5 teams which have the highest winning percentage
print(df.nlargest(5, 'Points')[['Team','Winning Percent']])
```

<font size="5">Group teams based on their “Best position” and print the sum of their points for all positions (5 points) Eg: Best Position Points

     1                   25000

```python
#Grouping based on BestPosition and summing their points for all positions and printing output
print(df.groupby(['BestPosition']).sum()[['Points']])
```
