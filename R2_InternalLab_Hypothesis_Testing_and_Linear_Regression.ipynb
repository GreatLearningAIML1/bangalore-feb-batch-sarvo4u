If you need to compare more than two data sets at a time, an ANOVA is your best bet. 

*The results from three experiments with overlapping 95% confidence intervals are given below, and we want to confirm that the results for all three experiments are not significantly different.*

#Assumption: All the 3 datasets (e1,e2 & e3) are random, independent, parametric & normally distributed
But before conducting ANOVA, test equality of variances (using Levene's test) is satisfied or not. If not, then mention that we cannot depend on the result of ANOVA

```python
import pandas as pd
e1 = pd.DataFrame([1.595440,1.419730,0.000000,0.000000])
e2 = pd.DataFrame([1.433800,2.079700,0.892139,2.384740])
e3 = pd.DataFrame([0.036930,0.938018,0.995956,1.006970])
```

Hint - You can use stats.levene function and stats.f_oneway function.

#### Answer:

```python
from scipy.stats import levene
# Null Hypothesis: Variances have equality
levene(e1,e2,e3)
#Since p_value>0.05 , so we do not reject Null hypothesis. Hence variances have equality. 
#We can depend on result on ANOVA
```

```python
# Null Hypothesis: Means are equal across all groups
from scipy.stats import f_oneway
f, p = f_oneway(e1,e2,e3)
print ('F value:', f)
print ('P value:', p, '\n')
#Answer3: Sin p_value > 0.05, we do not reject Null Hypothesis
#Mean are equal across all groups 
#And we want to confirm that the results for all three experiments are not significantly different.
```

# Question 6 
Get the binomial distribution with n = 10, p = .7 and k = [0, 1, 2, 3, 4, 5, 6, 7]

Hint - Use stats.binom.pmf() function for this.


```python
from scipy import stats
n, p = 10, 0.7
k = [0, 1, 2, 3, 4, 5, 6, 7]
binomial = stats.binom.pmf(k, n, p)
```

```python
print(binomial)
```

# Question 7 
Plot the distribution created in the above question.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(k,binomial,'o-')
plt.title('Binomial: n=%i , p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Number of Successes')
plt.ylabel('Probability of Successes')
plt.show()
```
