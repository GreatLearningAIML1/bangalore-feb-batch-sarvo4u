<font size="5">Question 2: Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? (10 points)</font>

```python
#μ1 - Mean of Old Scheme sample
#μ2 - Mean of New Scheme sample
#Null Hypothesis - H0 => μ1 >= μ2 i.e μ2 - μ1 <= 0 i.e Mean Sum Assured in New Scheme is equal to Mean Sum Assured 
#in Old Scheme at the 0.05 level of significance
#Alternate Hypothesis - H1 =>  μ2 > μ1 i.e μ2 - μ1 > 0 i.e Mean Sum Assured in New Scheme is greater than Mean Sum Assured 
#in Old Scheme at the 0.05 level of significance
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro, ttest_rel
t_statistic, p_value = ttest_rel(df1['New Scheme'],df1['Old Scheme'])
print(t_statistic, p_value)

```

```python
one_tailed_p_value = p_value/2
print(one_tailed_p_value)
```

<font size="5">Question 4b) What is the p- value of the hypothesis test if we test for a difference of $5000? Note: For Titan to break evenit has been calculated that the average output must increase by £5000 (Atleast)</font>

```python
#μ1 - Mean of Old Scheme sample
#μ2 - Mean of New Scheme sample
#μdiff - Mean of Difference of New Scheme sample and Old Scheme sample i.e. μdiff = μ2 - μ1
#Null Hyoothesis - μdiff <= 5000 i.e μdiff - 5000 <= 0
#Alternate Hypothesis - μdiff > 5000 i.e μdiff - 5000 > 0
#One tailed test (Right tailed)
t_statistic, p_value = ttest_1samp(df1['New Scheme'] - df1['Old Scheme'],5000)
print(t_statistic, p_value)
```

```python
#The p value above is two tailed p-value. One tailed p-value = p-value/2
oneTailedPValue = p_value /2
print('The p-value of the hypothesis test if we test for a difference of atleast $5000 to break even is',oneTailedPValue)
```

```python
#Since one tailed p-value is greater than 0.05, we DO NOT reject null hypothesis.
#Average output has not increased by $5000 to break even
```
